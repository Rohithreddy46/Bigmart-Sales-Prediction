```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=222529)
```

```python
from sklearn.impute import SimpleImputer
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor(random_state=42595)
```

```python
#rfr.fit(X_train, y_train)
```

```python
# Assuming X_train is a pandas DataFrame
imputer = SimpleImputer(strategy='mean')  # Or use 'median', 'most_frequent'
X_train_imputed = pd.DataFrame(imputer.fit_transform(X_train))

# Keep the column names
X_train_imputed.columns = X_train.columns

rfr = RandomForestRegressor(random_state=42595)
rfr.fit(X_train_imputed, y_train)  # Use the imputed data for training
```

```python
#y_pred = rfr.predict(X_test)
```

```python
# Assuming X_test is a pandas DataFrame and 'imputer' is already fitted
X_test_imputed = pd.DataFrame(imputer.transform(X_test))

# Keep the column names
X_test_imputed.columns = X_test.columns

y_pred = rfr.predict(X_test_imputed)  # Use the imputed test data for prediction
```

```python
y_pred.shape
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```
